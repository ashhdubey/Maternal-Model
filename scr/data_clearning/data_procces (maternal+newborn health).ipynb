```python
import pandas as pd
```

```python
df = pd.read_excel("/Users/ashutoshpdy/Developer/GitHub Project (git clone)/MAA/MAA_Maternal-Awareness-Aid/mlops/data/raw/Book2.xlsx", skiprows=1)
```

```python
df.columns = [
    "Name", "Age", "Gravida", "TT_Tika", "Pregnancy_Weeks",
    "Weight_kg", "Height_ft", "Blood_Pressure", "Anemia",
    "Jaundice", "Fetal_Position", "Fetal_Movement",
    "Fetal_HeartRate", "Urine_Albumin", "Urine_Sugar",
    "VDRL", "HBsAg", "Risk_Pregnancy"
]
```

```python
df["Weight_kg"] = df["Weight_kg"].str.replace(" kg", "", regex=False).astype(float)
df["Pregnancy_Weeks"] = df["Pregnancy_Weeks"].str.replace(" week", "", regex=False).astype(int)
```

```python
bp_split = df["Blood_Pressure"].str.split("/", expand=True)
df["BP_Systolic"] = pd.to_numeric(bp_split[0], errors="coerce")
df["BP_Diastolic"] = pd.to_numeric(bp_split[1], errors="coerce")
```

```python
df.drop(columns=["Blood_Pressure"], inplace=True)
```

```python
yes_no_cols = ["Urine_Albumin", "Urine_Sugar", "VDRL", "HBsAg", "Risk_Pregnancy"]
```

```python
for col in yes_no_cols:
    df[col] = df[col].map({"Yes": 1, "No": 0})
```

```python
df.to_csv(
    "/Users/ashutoshpdy/Developer/GitHub Project (git clone)/MAA/MAA_Maternal-Awareness-Aid/mlops/data/processed/anc_cleaned.csv",
    index=False
)
```
