```python
import pandas as pd
```

```python
df = pd.read_csv("/Users/ashutoshpdy/Developer/GitHub Project (git clone)/MAA/MAA_Maternal-Awareness-Aid/mlops/data/raw/Dataset - Updated.csv")
df.head(5)
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df["risk_level"] = df["risk_level"].str.lower()
```

```python
df["bmi"] = df["bmi"].fillna(df["bmi"].median())
```

```python
binary_cols = [
    "previous_complications",
    "preexisting_diabetes",
    "gestational_diabetes",
    "mental_health"
]
```

```python
df[binary_cols].isnull().sum()
```

```python
df[binary_cols] = df[binary_cols].fillna(0).astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna(subset=["risk_level"])
```

```python
numeric_cols = [
    "systolic_bp",
    "diastolic",
    "bs",
    "heart_rate"
]

for col in numeric_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
df.isna().sum()
```

```python
output_path = "/Users/ashutoshpdy/Developer/GitHub Project (git clone)/MAA/MAA_Maternal-Awareness-Aid/mlops/data/processed/Maternal Health Risk Data Set1.csv"
df.to_csv(output_path, index=False)
```
